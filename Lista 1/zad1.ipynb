```python
from matplotlib import pyplot as plt
import numpy as np

temp = 0
n = 1000
wartosci = []
numer = list(range(1,n+1))

p = np.random.uniform(size = n * 12)

for i in range(len(p)):
    
    temp = temp + p[i]

    if (i+1) % 12 == 0:
        wartosci.append(temp - 6)
        temp = 0


plt.hist(wartosci)
plt.title('Zmienne losowe otrzymane metodą tuzina')
plt.xlabel('Kolejna liczba losowa')
plt.ylabel('Wartość')
```

```python
from scipy.stats import norm

p = np.random.uniform(size = n)
p_inv = norm.ppf(p)

plt.hist(p_inv)
plt.title('Zmienne losowe otrzymane metodą odwrócpnej dystrybuanty')
plt.xlabel('Kolejna liczba losowa')
plt.ylabel('Wartość')

```

```python
U1 = np.random.uniform(size = 1000)
U2 = np.random.uniform(size = 1000)
R = np.sqrt(-2 * np.log(U1))
Theta = 2 * np.pi * U2
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

fig,(ax1,ax2) = plt.subplots(1,2)
temp = ax1.hist(X)
ax1.set_title("X")
temp = ax2.hist(Y)
ax2.set_title("Y")
```
